```python
import os
import sys
import warnings
import json
from datetime import datetime
import random
import numpy as np

import torch
import torch.nn as nn
import yaml

warnings.filterwarnings('ignore')

# Check if we're in Colab
try:
    import google.colab
    IN_COLAB = True
    print("Running in Google Colab")
except ImportError:
    IN_COLAB = False
    print("Not running in Colab")
```

```python
# Import project modules
from models.mobilevit import mobilevit_s
from data.dataset import create_data_loaders
from training.trainer import MobileViTTrainer, plot_training_history
from utils.utils import print_model_summary, setup_device_and_memory, setup_logging

print("Project modules imported successfully")
```

```python
# Dataset configuration for Kaggle dataset in Colab
DATASET_PATH = "/root/.cache/kagglehub/datasets/shivamardeshna/real-and-fake-images-dataset-for-image-forensics/versions/1"

# note that directory structure is Data Set 1/Data Set 1/train, not Data Set 1/train;
# the above structure is similar for validation and test and for all other datasets in this directory

# Verify dataset path exists
if os.path.exists(DATASET_PATH):
    print(f"Dataset found at: {DATASET_PATH}")
    print(f"Dataset contents: {os.listdir(DATASET_PATH)}")

    # Explore the structure of each Data Set directory
    for data_set in sorted(os.listdir(DATASET_PATH)):
        data_set_path = os.path.join(DATASET_PATH, data_set)
        data_set_path = os.path.join(data_set_path, data_set)
        if os.path.isdir(data_set_path):
            print(f"\n{data_set} contents: {os.listdir(data_set_path)}")

            # Check for common deepfake dataset patterns
            for subdir in os.listdir(data_set_path):
                subdir_path = os.path.join(data_set_path, subdir)
                contents = []
                for sub_subdir in os.listdir(subdir_path):
                    sub_subdir_path = os.path.join(subdir_path, sub_subdir)
                    if os.path.isdir(sub_subdir_path):
                        file_count = len([f for f in os.listdir(sub_subdir_path)
                                        if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
                        contents.append(f"{sub_subdir}: {file_count} images")
                print(f"   {subdir} contents: {contents}")
else:
    print(f"Dataset not found at: {DATASET_PATH}")
    print("Please ensure the dataset is downloaded using kagglehub")

# For this dataset, we'll use Data Set 1 as our main dataset
# You can change this to any of the available Data Sets
SELECTED_DATASET = "Data Set 1"
ACTUAL_DATASET_PATH = os.path.join(DATASET_PATH, SELECTED_DATASET)
ACTUAL_DATASET_PATH = os.path.join(ACTUAL_DATASET_PATH, SELECTED_DATASET)

print(f"\nUsing dataset: {SELECTED_DATASET}")
print(f"Selected dataset path: {ACTUAL_DATASET_PATH}")

# Configuration dictionary
config = {
    # Data configuration - using the selected dataset
    'data_dir': ACTUAL_DATASET_PATH,
    'batch_size': 32,
    'image_size': 224,
    'augmentation': 'advanced',

    # Model configuration
    'model_name': 'mobilevit_s',
    'num_classes': 2,
    'pretrained': True,

    # Training configuration
    'epochs': 50,  # Reduced for Colab
    'learning_rate': 1e-4,
    'weight_decay': 1e-2,
    'mixed_precision': True,
    'gradient_clipping': 1.0,
    'patience': 10,

    # System configuration
    'num_workers': 2,  # Reduced for Colab
    'seed': 42,

    # Logging configuration
    'experiment_name': f"mobilevit_deepfake_{datetime.now().strftime('%Y%m%d_%H%M%S')}",
    'log_dir': '/content/runs',
    'checkpoint_dir': '/content/checkpoints',
    'results_dir': '/content/results'
}

print(f"Experiment: {config['experiment_name']}")
print(f"Epochs: {config['epochs']}")
print(f"Batch size: {config['batch_size']}")
print(f"Using data directory: {config['data_dir']}")
print(f"Using model: {config['model_name']}")
print(f"Using augmentation: {config['augmentation']}")
```

```python
# Setup directories (using utility from main.py)
def setup_directories(config: dict):
    directories = ['log_dir', 'checkpoint_dir', 'results_dir']
    for dir_key in directories:
        if dir_key in config:
            os.makedirs(config[dir_key], exist_ok=True)
            print(f"Created directory: {config[dir_key]}")

def set_random_seeds(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    print(f"Random seed set to: {seed}")

# Setup environment
setup_directories(config)
set_random_seeds(config['seed'])
device = setup_device_and_memory()
print(f"Using device: {device}")
```

```python
# Create model using imported module
model = mobilevit_s(num_classes=config['num_classes'], pretrained=config['pretrained'])
model = model.to(device)

# Print model summary
input_size = (3, config['image_size'], config['image_size'])
print_model_summary(model, input_size=input_size)

print("Model created and moved to device")
```

```python
# Determine data directories
train_dir = os.path.join(config['data_dir'], 'train')
val_dir = os.path.join(config['data_dir'], 'validation')
test_dir = os.path.join(config['data_dir'], 'test')

# If no separate train/val dirs, use the main directory
if not os.path.exists(train_dir) or not os.path.exists(val_dir):
    print("No separate train/val directories found. Using main directory with split.")
    train_dir = config['data_dir']
    val_dir = config['data_dir']
    test_dir = None # No separate test directory if no train/val

print(f"Dataset directory: {config['data_dir']}")
print(f"Train directory: {train_dir}")
print(f"Validation directory: {val_dir}")
if test_dir:
    print(f"Test directory: {test_dir}")

# Create data loaders using imported function
try:
    train_loader, val_loader, test_loader = create_data_loaders(
        train_dir=train_dir,
        val_dir=val_dir,
        test_dir=test_dir,
        batch_size=config['batch_size'],
        image_size=config['image_size'],
        num_workers=config['num_workers'],
        augmentation_type=config['augmentation']
    )
    print("Data loaders created successfully")
except Exception as e:
    print(f"Error creating data loaders: {e}")
    print("Please check the dataset path and structure")
```

```python
# Create trainer using imported class
trainer = MobileViTTrainer(model, train_loader, val_loader, config)
print("Trainer initialized with imported MobileViTTrainer class")

# Setup logging
logger = setup_logging(config['log_dir'])
logger.info(f"Starting experiment: {config['experiment_name']}")
logger.info(f"Configuration: {config}")
```

```python
# Start training
print("" + "=" * 60)
print("STARTING TRAINING")
print("=" * 60)
print(f"Device: {device}")
print(f"Mixed precision: {config['mixed_precision']}")
print(f"Batch size: {config['batch_size']}")
print(f"Learning rate: {config['learning_rate']}")
print(f"Epochs: {config['epochs']}")

# Train the model using imported trainer
training_history = trainer.train()

print("\n" + "=" * 60)
print("TRAINING COMPLETED")
print("=" * 60)
print(f"Best validation F1 score: {trainer.best_f1:.4f}")
```

```python
# Plot training history using imported function
history_plot_path = os.path.join(config['results_dir'], 'training_history.png')
plot_training_history(training_history, save_path=history_plot_path)
print(f"Training history plot saved to: {history_plot_path}")

# Display the plot
from IPython.display import Image, display
if os.path.exists(history_plot_path):
    display(Image(history_plot_path))
```

```python
# Evaluate on test set if available using imported trainer method
if test_loader:
    print("\n" + "=" * 60)
    print("EVALUATING ON TEST SET")
    print("=" * 60)

    test_metrics = trainer.evaluate(test_loader, save_results=True)
    logger.info(f"Final test metrics: {test_metrics}")

    print("\n" + "="*60)
    print("FINAL TEST RESULTS")
    print("="*60)
    print(f"Test Accuracy: {test_metrics.get('accuracy', 'N/A'):.4f}")
    print(f"Test Precision: {test_metrics.get('precision', 'N/A'):.4f}")
    print(f"Test Recall: {test_metrics.get('recall', 'N/A'):.4f}")
    print(f"Test F1 Score: {test_metrics.get('f1', 'N/A'):.4f}")
    print("="*60)
else:
    print("No test data available. Using validation metrics as final results.")
    test_metrics = None

# Save final results summary
summary = {
    'experiment_name': config['experiment_name'],
    'best_validation_f1': trainer.best_f1,
    'test_metrics': test_metrics,
    'config': config
}

summary_path = os.path.join(config['results_dir'], 'experiment_summary.json')
with open(summary_path, 'w') as f:
    json.dump(summary, f, indent=2)

print(f"\nExperiment summary saved to: {summary_path}")

# Log completion
logger.info("Experiment completed successfully")
logger.info(f"Best validation F1: {trainer.best_f1:.4f}")
if test_metrics:
    logger.info(f"Test F1: {test_metrics.get('f1', 'N/A'):.4f}")
```

```python
# Compress results for download
import zipfile

def create_results_zip():
    zip_path = '/content/deepfake_detection_results.zip'

    with zipfile.ZipFile(zip_path, 'w') as zipf:
        # Add results directory
        for root, dirs, files in os.walk(config['results_dir']):
            for file in files:
                file_path = os.path.join(root, file)
                arc_name = os.path.relpath(file_path, '/content')
                zipf.write(file_path, arc_name)

        # Add best model checkpoint
        best_model_path = os.path.join(config['checkpoint_dir'], 'best_model.pth')
        if os.path.exists(best_model_path):
            arc_name = os.path.relpath(best_model_path, '/content')
            zipf.write(best_model_path, arc_name)

    print(f"Results compressed to: {zip_path}")
    return zip_path

# Create zip file
zip_path = create_results_zip()

# Download in Colab
if IN_COLAB:
    from google.colab import files
    print("Downloading results...")
    files.download(zip_path)
else:
    print(f"Results available at: {zip_path}")

print("\n" + "="*60)
print("TRAINING COMPLETED SUCCESSFULLY!")
print(f"Best Validation F1 Score: {trainer.best_f1:.4f}")
if test_metrics:
    print(f"Test F1 Score: {test_metrics.get('f1', 'N/A'):.4f}")
print(f"Experiment: {config['experiment_name']}")
print("="*60)
```
